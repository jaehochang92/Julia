### Dot(`.`) Broadcasting

```julia
countries == "US"
```

```julia
countries .== "US"
```

```julia
US_row = findfirst(countries .== "US")
```

```julia
US_df_row = df[US_row, :]
```

```julia
US_v = convert(Vector, US_df_row[5:end])
```

```julia
Pkg.add("Plots")
using Plots
```

```julia
plot(US_v)
```

```julia
dates = String.(names(df)[5:end])  # dot broadcasting
dates = names(df)[5:end]  # same
```

```julia
dates[1]
```

```julia
Pkg.add("Dates")
using Dates
```

```julia
parse(Date, dates[1], Dates.DateFormat("m/d/y"))
```

```julia
dates = parse.(Date, dates, Dates.DateFormat("m/d/y")) .+ Year(2e3)
```

```julia
plot(dates, US_v, xticks=dates[1:30:end], xrotation=46, leg=:topleft, label="US data",
m=:o)
xlabel!("date")
ylabel!("confirmed cases in US")
title!("Confirmed COVID-19 cases")

annotate!(dates[end], US_v[end], text("US", :blue, :left))
```

```julia
plot(dates, US_v, xticks=dates[1:30:end], xrotation=46, leg=:topleft, label="US data",
m=:o, yscale=:log10)
xlabel!("date")
ylabel!("confirmed cases in US")
title!("Confirmed COVID-19 cases")
```
