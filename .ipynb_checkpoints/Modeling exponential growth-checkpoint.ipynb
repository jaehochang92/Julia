### Modeling infection spreading through campus

**Assumptions:**
> No influence from outside\
> No births or deaths\
> Virus is transmitted by proximity\
> Stay infected once infected

**Model**

$n$: days ($0, \cdots, T$ for a $T<\infty$)

$I_n$: # of infectious people on $n$th day

$S_n$: # of susceptible people on $n$th dat

Suppose that $N:=I_n+S_n$ is a population and constant

Model Candidates are...
1. $I_{n+1}=f(I_n)$: deterministic model
1. $I_{n+1}=f(I_n)+\xi$: stochastic model with random noises $\xi$

Suppose that...
* Each individual infects $c$ other people each day.

    $I_{n+1}=(1+c)I_{n}$

> Initialize as\
> $n=0,~I_0=1,~S_0=N-1$

> $c$ is a growth rate

```julia
I₀ = 1  # I\_0 + TAB
```

```julia
c = .01
```

```julia
λ = 1 + c  # unicode
```

```julia
T = 10
# 𝐈 = zeros(Int64, T)
𝐈 = zeros(Float64, T)
```

```julia
𝐈[1] = I₀
for n in 1:T-1
    𝐈[n + 1] = λ * 𝐈[n]
    @show n + 1, 𝐈[n]
end
```

```julia
using Plots
```

```julia
plot(𝐈, m=:o, label="I[n]", legend=:topleft)
```

```julia
function run_infsc(T=20, λ=1.1, I₀=1)
    𝐈 = zeros(Float64, T)
    𝐈[1] = I₀
    
    for n in 1:T-1
        𝐈[n + 1] = λ * 𝐈[n]
#         @show n + 1, 𝐈[n]
    end

    return 𝐈
end
```

```julia
sim = run_infsc(50);
plot(sim, m=:o, label="I[n]", legend=:topleft)
```
